```python
#Импорт

import numpy as np
from matplotlib import pylab as plt
import math
from scipy.linalg import solve
```

```python
#
x = np.arange(1,15,0.1)
y = np.sin(x/5) * np.exp(x/10) + 5*np.exp(-x/2)

plt.plot(x, y)
plt.show()
```

```python
#w0 + w1*x = b
x1 = 1.
x2 = 15.
x = np.array([x1,x2])
print 'x',x

b1 = np.sin(x1/5) * np.exp(x1/10) + 5*np.exp(-x1/2)
b2 = np.sin(x2/5) * np.exp(x2/10) + 5*np.exp(-x2/2)
b = np.array([b1,b2])
print 'b',b

a = np.array([[1,x1],[1,x2]])
print 'a',a

w = solve(a,b)
print 'w',w

plt.plot(x,b,'r')

plt.plot()
plt.show()

```

```python
#w0 + w1*x + w2*x**2 = b
x1 = 1.
x2 = 8.
x3 = 15.
x = np.array([x1,x2,x3])
print 'x',x

b1 = np.sin(x1/5) * np.exp(x1/10) + 5*np.exp(-x1/2)
b2 = np.sin(x2/5) * np.exp(x2/10) + 5*np.exp(-x2/2)
b3 = np.sin(x3/5) * np.exp(x3/10) + 5*np.exp(-x3/2)
b = np.array([b1,b2,b3])
print 'b',b

a = np.array([[1,x1,x1**2],[1,x2,x2**2],[1,x3,x3**2]])
print 'a',a

w = solve(a,b)
print 'w',w

plt.plot(x,b,'b')
plt.show()
```

```python
#w0 + w1*x + w2*x**2 + w3*x**3= b
x1 = 1.
x2 = 4.
x3 = 10.
x4 = 15.
x = np.array([x1,x2,x3,x4])
print 'x',x

b1 = np.sin(x1/5) * np.exp(x1/10) + 5*np.exp(-x1/2)
b2 = np.sin(x2/5) * np.exp(x2/10) + 5*np.exp(-x2/2)
b3 = np.sin(x3/5) * np.exp(x3/10) + 5*np.exp(-x3/2)
b4 = np.sin(x4/5) * np.exp(x4/10) + 5*np.exp(-x4/2)
b = np.array([b1,b2,b3,b4])
print 'b',b

a = np.array([[1,1,1,1],[1,4,16,64],[1,10,100,1000],[1,15,225,3375]])
print 'a',a

w = solve(a,b)
print 'w',round(w[0],2),round(w[1],2),round(w[2],2),round(w[3],2)

plt.plot(x,b,'b')
plt.show()
```
