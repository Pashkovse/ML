# Анализ временных рядов

Алгоритм подбора модели в классе ARIMA состоит в следующем:
* В первую очередь необходимо построить график ряда и посмотреть на него. Уже из визуального ана- лиза можно сделать определённые выводы: есть ли в данных сезонность, какой сезонный период, есть ли в ряде пропуски и выбросы, необходимо ли стабилизировать дисперсию, стоит ли исключить из рассмотрения начало ряда, потому что значения в начале совсем не похожи на значения в конце.
* Следующий шаг — это стабилизация дисперсии при необходимости. Стабилизация производится с по- мощью метода Бокса-Кокса или логарифмированием, что является частным случаем того же метода.
* Если исследуемый ряд нестационарен, необходимо подобрать порядок дифференцирования, при котором он становится стационарным. Таким образом фиксируются параметры d, D модели ARIMA.
* Далее необходимо построить графики автокорреляционной функции (ACF - корреляция между исходным рядом и сдвинутым на тау) и частичной автокорреля- ционной функции (PACF - автокорреляция после снятия авторегрессии предыдущего шага). Из ACF определить q (номер последнего значимого несезаонного лага), Q (номер последнего значимого сезонного лага), Из PACF по аналогии - p,P.
* Полученные модели необходимо обучить, сравнить их по информационному критерию Акаике и выбрать ту, которая его минимизирует. AIC = −2 ln L + 2k, где L — правдоподобие, k = P + Q + p + q + 1 — число параметров в модели.
* Необходимо посмотреть на остатки получившейся модели, чтобы понять, насколько хорошей она получилась, можно ли, теоретически, её улучшить, нет ли в ней каких-то видимых недостатков.

```python
import pandas as pd
%pylab inline
```

Временной ряд ежемесячного производства молока

```python
milk = pd.read_csv('monthly-milk-production.csv',';', index_col=['month'], parse_dates=['month'], dayfirst=True)
milk.milk.plot()
```

```python
milk.head()
```

Проверка на стационарность (критерий Дики-Фуллера)

```python
import statsmodels.api as sm
print ('p_val ', sm.tsa.stattools.adfuller(milk.milk)[1])
```

Построим график ежедневного показателя

```python
milk['milk_scale'] = milk.milk / milk.index.days_in_month
milk['milk_scale'].plot()
```

Сделаем его стационарным с помощью дифференцирования

```python
milk.daily_diff1 = milk.milk_scale - milk.milk_scale.shift(1)
milk.daily_diff1.plot()
```

```python
milk.daily_diff12 = milk.daily_diff1 - milk.daily_diff1.shift(12)
milk.daily_diff12.plot()
```

Построим график автокорреляционной и частично автокорреляционной функций

```python
import statsmodels.api as sm
sm.graphics.tsa.plot_acf(milk.daily_diff12.dropna().values.squeeze(), lags=50)
```

```python
sm.graphics.tsa.plot_pacf(milk.daily_diff12.dropna().values.squeeze(), lags=50)
```
